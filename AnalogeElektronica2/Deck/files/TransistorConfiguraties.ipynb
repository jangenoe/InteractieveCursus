### De MOS transistor geschakeld als 3-terminal component

We catalogeren de transistors waarvan de 3 terminalen gebruikt worden in 3 groepen, bepaald door welk contact aan een vaste spanning ligt:


- gemeenschappelijke source
  - de ingang is dan verbonden met de gate (=spanningsbron )
  - de uitgang is dan verbonden met de drain
  - deze transistor werkt als een spanningsversterker (amplifier)
- gemeenschappelijke  drain
  - de ingang is dan verbonden met de gate (=spanningsbron)
  - de uitgang is dan verbonden met de source
  - de spanning aan de source volgt de spanning aan de ingang (source follower)
- gemeenslijke gate
  - de ingang is dan verbonden met de source (stroombron)
  - de uitgang is dan venbonden met de drain
  - de spanning aan de drain wordt bepaald door de stroom aan drain (cascode trap)

Elk van deze 3-terminal configuraties bespreken we hieronder in meer detail

### Gemeenschappelijke Source Schakeling: GSS

```python
with schemdraw.Drawing():
    T = NFet(reverse=True)
    Ground(xy=T.source)
    Line( d='left',  xy=T.gate, l=0.75)
    Dot(open=True,label='$V_{in}$')
    Dot(xy=T.drain)
    Line( d='right', l=0.75)
    Dot(open=True,label='$V_{uit}$')
    RL = Resistor(d='up',xy=T.drain,label='$R_L$')
    Vdd(label='$V_{dd}$')
    CurrentLabel(color='green',xy=(RL.start[0]+0.6,RL.start[1]+1.5)).reverse().label(r'$I_{uit}=g_{m} V_{in}$',color='green',ofst=(0,-2.25))
    flow.Box(w=6,color='blue',fill='lightblue').label(r'$V_{uit}=V_{dd}- R_L g_{m} V_{in}$', fontsize=18).at((RL.start[0]+7,RL.start[1]));
```

## Bipolaire transistor configuraties

### Gemeenschappelijke Emitter Schakeling: GES

Als voor een bipolaire transistor de emitter de gemeenschappelijke klem is, is de ingang steeds de basis van bipolaire transistor en de uitgang steeds de collector van de bipolaire tranistor. De ingang kan in principe zowel in stroom (zie {numref}`GES`) als in spanning aangestuurd worden (zie {numref}`GES2`).

In deze configuratie is het doel van de transistor in eerste instantie om spanningsversterking op te leveren. {numref}`GESdefA` toont de spanningsversterking die kan bekomen worden.


```python
with schemdraw.Drawing():
    T = BjtNpn()
    Ground(xy=T.emitter)
    RI = Resistor(d='left',l=1,  xy=T.base,label='$R_{in}$')
    Line( d='left', l=0.75)
    Dot(open=True,label='$V_{in}$')
    Dot(xy=T.collector)
    Line( d='right', l=0.75)
    Dot(open=True,label='$V_{uit}$')
    RL = Resistor(d='up',xy=T.collector,label='$R_L$')
    Vdd(label='$V_{dd}$')
    LoopArrow(xy=(RI.start[0]-1,RI.start[1]-1), color='r',direction='cw',theta1=0,theta2=120,width=1.5).label('$I_{in}$')
    CurrentLabel(color='green',xy=(RL.start[0]+1,RL.start[1]+1.5)).reverse().label(r'$I_{uit}=\beta_{F} I_{in}$',color='green',ofst=(0,-2))
    flow.Box(w=6,color='blue',fill='lightblue').label(r'$V_{uit}=V_{dd}- R_L \beta_{F} I_{in}$', fontsize=18).at((RL.start[0]+6,RL.start[1]+1));
```
