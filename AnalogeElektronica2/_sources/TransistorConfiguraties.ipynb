## Tweepoort



```python
from cursusdef import *
from schemdraw import flow
```

Wanneer we een ingangsspanning $V_{in}$ aanleggen aan een transistor die gebruikt wordt als tweepoorten en een uitgangsspanning $V_{uit}$ bekomen, hebben deze spanningen deze enkel zin ten opzichte van een referentie. In dit overzicht van de transistor configuraties nemen we deze referentie gemeenschappelijk voor de ingang en de uitgang zoals je kan zien in figuur {numref}`TdefA`. De kleinsignaalversterking $A$ definieren we dan vervolgens als de verhouding tussen differentiele uitgangsspanning ($\partial V_{uit}$) en de differentiele ingangsspanning ($\partial V_{in}$).

```python
with schemdraw.Drawing():
    idot = Dot(open=True)
    Line(d='right',l=1)
    poort3=Ic(pins=[IcPin( side='left', anchorname='l'),IcPin(side='bottom', anchorname='b'),IcPin(side='right', anchorname='r')],
            edgepadW = .5,  # Make it a bit wider
            pinspacing=1).label('3 terminal\n component', 'center', fontsize=12).anchor('l')
    gdot=Dot(xy=poort3.b)
    Line(xy=poort3.b,tox=idot.start)
    gdot1 =Dot(open=True)
    Line(xy=poort3.r,d='right',l=1)
    odot = Dot(open=True)
    Line(xy=poort3.b,tox=odot.start)
    gdot2 =Dot(open=True)
    Gap(d='up', xy= gdot1.start, toy= idot.end, color='r', label=['-','$v_{in}$','$+$'],lblofst=0.5)
    Gap(d='up', xy= gdot2.start, toy= odot.end, color='g', label=['-','$v_{uit}$','$+$'],lblofst=-0.5)
    flow.Box(w=3,color='blue',fill='lightblue').label(r'$A=\frac{\partial v_{uit}}{\partial v_{in}}$', fontsize=18).at((odot.start[0]+3,odot.start[1]+1));
```

Wanneer we een ingangsstroom $I_{in}$ aanleggen aan een transistor die gebruikt wordt als tweepoort en een uitgangsstroom $I_{uit}$ bekomen, moeten deze stroom op een of andere manier ook terugstromen. In dit overzicht van de transistor configuraties nemen we de gemeenschappelijk klem voor de ingang en de uitgang als de klem waarlangs de stroom terugvloeit, zoals je kan zien in figuur {numref}`TdefB`. De kleinsignaalstroomversterking $\beta$ definieren we dan vervolgens als de verhouding tussen differentiele uitgangsstroom ($\partial I_{uit}$) en de differentiele ingangsstroom ($\partial I_{in}$).

```python
with schemdraw.Drawing():
    idot = Dot(open=True)
    Line(d='right',l=1)
    poort3=Ic(pins=[IcPin( side='left', anchorname='l'),IcPin(side='bottom', anchorname='b'),IcPin(side='right', anchorname='r')],
            edgepadW = .5,  # Make it a bit wider
            pinspacing=1).label('3 terminal\n component', 'center', fontsize=12).anchor('l')
    gdot=Dot(xy=poort3.b)
    Line(xy=poort3.b,tox=idot.start)
    gdot1 =Dot(open=True)
    Line(xy=poort3.r,d='right',l=1)
    odot = Dot(open=True)
    Line(xy=poort3.b,tox=odot.start)
    gdot2 =Dot(open=True)
    LoopArrow(xy=(idot.start+(gdot.start+gdot1.start)/2)/2, color='r', direction='cw',theta1=-120,theta2=120,width=1.5).label('$I_{in}$')
    LoopArrow(xy=(odot.start+(gdot.start+gdot2.start)/2)/2, color='g', direction='ccw',theta1=60,theta2=-60,width=1.5).label('$I_{uit}$')
    flow.Box(w=3,color='blue',fill='lightblue').label(r'$\beta=\frac{\partial I_{uit}}{\partial I_{in}}$', fontsize=18).at((odot.start[0]+3,odot.start[1]+1));
```

Op een gelijkaardige wijze definieren we de transconductantie $g_m$ als de verhouding tussen differentiele uitgangsstroom ($\partial I_{uit}$) en de differentiele ingangsspanning ($\partial V_{in}$).

```python
with schemdraw.Drawing():
    idot = Dot(open=True)
    Line(d='right',l=1)
    poort3=Ic(pins=[IcPin( side='left', anchorname='l'),IcPin(side='bottom', anchorname='b'),IcPin(side='right', anchorname='r')],
            edgepadW = .5,  # Make it a bit wider
            pinspacing=1).label('3 terminal\n component', 'center', fontsize=12).anchor('l')
    gdot=Dot(xy=poort3.b)
    Line(xy=poort3.b,tox=idot.start)
    gdot1 =Dot(open=True)
    Line(xy=poort3.r,d='right',l=1)
    odot = Dot(open=True)
    Line(xy=poort3.b,tox=odot.start)
    gdot2 =Dot(open=True)
    LoopArrow(xy=(odot.start+(gdot.start+gdot2.start)/2)/2, color='g', direction='ccw',theta1=60,theta2=-60,width=1.5).label('$I_{uit}$')
    Gap(d='up', xy= gdot1.start, toy= idot.end, color='r', label=['-','$v_{in}$','$+$'],lblofst=0.5)
    flow.Box(w=3,color='blue',fill='lightblue').label(r'$g_m=\frac{\partial I_{uit}}{\partial v_{in}}$', fontsize=18).at((odot.start[0]+3,odot.start[1]+1));
```

## Eenpoort

We kunnen een transistor ook als een tweeterminal component gebruiken. Hierdoor bekomen we een eenpoort.
We leggen een spanning  aan over de component en krijgen er een stroom door, of omgekeerd. Er is extern geen derde contact dat een controle van de stroom toelaat. Dit kan bekomen worden door 2 van de 3 terminals kort te sluiten of er een vaste spanning over aan te brengen.

```python
with schemdraw.Drawing():
    idot = Dot(open=True)
    Line(d='right',l=1)
    poort2=Ic(pins=[IcPin( side='left', anchorname='l'),IcPin(side='bottom', anchorname='b')],
            edgepadW = .5,  # Make it a bit wider
            pinspacing=1).label('2 terminal\n component', 'center', fontsize=12).anchor('l')
    Line(xy=poort2.b,tox=idot.start)
    gdot =Dot(open=True)
    LoopArrow(xy=(idot.start+(gdot.start+poort2.b)/2)/2, color='r',direction='cw',theta1=-120,theta2=120,width=1.5).label('$I_{in}$')
    Gap(d='up', xy= gdot1.start, toy= idot.end, color='r',label=['-','$v_{in}$','$+$'],lblofst=0.5)  
    flow.Box(w=3,color='blue',fill='lightblue').label(r'$g_o=\frac{\partial I_{uit}}{\partial v_{in}}$', fontsize=18).at((gdot.start[0]+6,gdot.start[1]+1));
    flow.Box(w=3,color='blue',fill='lightblue').label(r'$R_o=\frac{\partial v_{in}}{\partial I_{uit}}$', fontsize=18).at((gdot.start[0]+10,gdot.start[1]+1));
```

## MOS transistor configuraties

### De MOS transistor geschakeld als 2-terminal component

Ondanks het feit dat een MOS transistor een 3-terminal component is, kunnen we deze in veel schakelingen als een 2-terminal component gebruiken. 

- De MOS transistor geschakeld als diode
- De MOS transistor geschakeld als actieve belasting

We werken hieronder beide configuraties in detail uit.


#### De MOS transistor geschakeld als diode

Een diode is een twee-terminal component. We leggen er een spanning over aan en krijgen er een stroom door, of omgekeerd. Er is geen derde contact dat een controle van de stroom toelaat. Een transistor daarentegen heeft wel dat derde contact dat toelaat de stroom te controleren. Wanneer nu echter de gate van de transistor met de drain verbonden wordt (en de bulk met de source) hebben we slechts 2 externe contacten meer over. We bekomen hieruit een stroom-spanningskarakteristiek die kwadratisch is (dus niet logaritmisch zoals bij de diode).

:::{warning}
dit is de transistor als diode geschakeld, maar daarom is het nog geen diode.
:::


```python
with schemdraw.Drawing():
    T = NFet(bulk=True)
    Line(xy=T.bulk,l=0.5,d='left')
    Line(toy=T.source,d='down')
    Line(tox=T.source,d='right')
    Dot()
    Ground(xy=T.source)
    Line( d='up',  xy=T.gate, toy=T.drain)
    Line( d='left',   tox=T.drain)
    Dot()
    Line( d='up',  l=0.5)
    Dot(open=True,label='$V_{in}$')
    CurrentLabel(color='r',xy=(T.drain[0]-0.3,T.drain[1]+0.5)).reverse().label(r'$I_{in}$',color='r')
    flow.Box(w=6,color='blue',fill='lightblue').label(r'$I_{in}=\frac{\mu C_{ox}}{2}\frac{W}{L}(V_{in}-V_{T})^2$', fontsize=18).at((T.drain[0]+6,T.drain[1]));
    flow.Box(w=8,color='blue',fill='lightblue').label(r'$g_o=\frac{\partial I_{uit}}{\partial v_{in}}=\mu C_{ox}\frac{W}{L}(V_{in}-V_{T})$', fontsize=18).at((T.drain[0]+6,T.drain[1]-3));
```

#### De MOS transistor als actieve belasting



### De MOS transistor geschakeld als 3-terminal component

We catalogeren de transistors waarvan de 3 terminalen gebruikt worden in 3 groepen, bepaald door welk contact aan een vaste spanning ligt:


- gemeenschappelijke source
  - de ingang is dan verbonden met de gate (=spanningsbron )
  - de uitgang is dan verbonden met de drain
  - deze transistor werkt als een spanningsversterker (amplifier)
- gemeenschappelijke  drain
  - de ingang is dan verbonden met de gate (=spanningsbron)
  - de uitgang is dan verbonden met de source
  - de spanning aan de source volgt de spanning aan de ingang (source follower)
- gemeenslijke gate
  - de ingang is dan verbonden met de source (stroombron)
  - de uitgang is dan venbonden met de drain
  - de spanning aan de drain wordt bepaald door de stroom aan drain (cascode trap)

Elk van deze 3-terminal configuraties bespreken we hieronder in meer detail

#### Gemeenschappelijke Source Schakeling: GSS

```python
with schemdraw.Drawing():
    T = NFet(reverse=True,bulk=True)
    Line(xy=T.bulk,l=0.5,d='right')
    Line(toy=T.source,d='down')
    Line(tox=T.source,d='left')
    Ground(xy=T.source)
    Line( d='left',  xy=T.gate, l=0.75)
    Dot(open=True,label='$V_{in}$')
    Dot(xy=T.drain)
    Line( d='right', l=0.75)
    Dot(open=True,label='$V_{uit}$')
    RL = Resistor(d='up',xy=T.drain,label='$R_L$')
    Vdd(label='$V_{dd}$')
    CurrentLabel(color='green',xy=(RL.start[0]+0.6,RL.start[1]+1.5)).reverse().label(r'$I_{uit}=g_{m} V_{in}$',color='green',ofst=(0,-2.25))
    flow.Box(w=6,color='blue',fill='lightblue').label(r'$V_{uit}=V_{dd}- R_L g_{m} V_{in}$', fontsize=18).at((RL.start[0]+7,RL.start[1]));
```

## Bipolaire transistor configuraties

### Gemeenschappelijke Emitter Schakeling: GES

Als voor een bipolaire transistor de emitter de gemeenschappelijke klem is, is de ingang steeds de basis van bipolaire transistor en de uitgang steeds de collector van de bipolaire tranistor. De ingang kan in principe zowel in stroom (zie {numref}`GES`) als in spanning aangestuurd worden (zie {numref}`GES2`).

In deze configuratie is het doel van de transistor in eerste instantie om spanningsversterking op te leveren. {numref}`GESdefA` toont de spanningsversterking die kan bekomen worden.
