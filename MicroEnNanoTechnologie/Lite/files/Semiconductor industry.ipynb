# Semiconductor industry

Deze pagina overloopt alle fabs in de wereld waar er momenteel chips worden gemaakt, op voorwaarde dat deze fab gekend is op [wikipedia](https://en.wikipedia.org/wiki/List_of_semiconductor_fabrication_plants).


```python
from wikitables import import_tables
semiconductorFabs = import_tables('List_of_semiconductor_fabrication_plants')
```

```python
print('Er zijn momenteel ',len(semiconductorFabs[0].rows), 'werkende chip fabs gekend.')
```
