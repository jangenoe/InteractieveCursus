# Technology nodes



```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle


def nmos(ax,L=350,sourceimplant=100,bulkimplant=700,SL=500,DL=500,GSO=20,GDO=20,tox=50,gatethickness=300,xoffset=0):
    ax.add_patch( Rectangle((xoffset-L-2*SL,-bulkimplant),2*(SL+L+DL),bulkimplant,fc ='green', ec ='green',alpha=0.15,lw = 1) )
    ax.add_patch( Rectangle((xoffset-L/2-SL,-sourceimplant),SL,sourceimplant,fc ='green', ec ='green',alpha=0.4,lw = 1) )
    ax.add_patch( Rectangle((xoffset+L/2,-sourceimplant),DL,sourceimplant,fc ='green', ec ='green',alpha=0.4,lw = 1) )
    ax.add_patch( Rectangle((xoffset-L/2-GSO,tox),L+GSO+GDO,gatethickness,fc ='red', ec ='red',alpha=0.4,lw = 1) )

fig, ax = plt.subplots(1, 1, figsize=(10, 5))
nmos(ax=ax)
ax.set_xlim(-700,700)
ax.set_xlabel("Channel direction (nm)")
ax.set_ylim(-250,450)
ax.set_ylabel("Depth profile (nm)");
```
